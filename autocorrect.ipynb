```python
from spellchecker import SpellChecker
import string

spell = SpellChecker()
```

```python
def autocorrect_word(word):

    prefix = ''
    suffix = ''
    import string

    while word and word[0] in string.punctuation:
        prefix += word[0]
        word = word[1:]
    while word and word[-1] in string.punctuation:
        suffix = word[-1] + suffix
        word = word[:-1]

    if not word:
        return prefix + suffix

    lower_word = word.lower()


    if lower_word in custom_dict:
        corrected = custom_dict[lower_word]
    else:

        corrected = spell.correction(lower_word) or word


    if word[0].isupper():
        corrected = corrected.capitalize()

    return prefix + corrected + suffix
```

```python
def autocorrect_sentence(sentence):
    words = sentence.split()
    corrected = [autocorrect_word(w) for w in words]
    return " ".join(corrected)
```

```python

custom_dict = {
    "ths": "this",
    "autocorret": "autocorrect",
    "writen": "written",
    "hav": "have",
    "hop": "hope",
    "bt": "but",
    "paragrph": "paragraph",
    "gras":"grass"
}
```

```python
paragraph = input("Paste your paragraph:\n\n")
print("\n\nthe corrected output is:\n\n\n",autocorrect_sentence(paragraph))
```

```python
# Colors
RED = "\033[91m"
GREEN = "\033[92m"
RESET = "\033[0m"

def color_compare(original, corrected):
    """Color incorrect and corrected words."""
    orig_words = original.split()
    corr_words = corrected.split()

    result = []

    for o, c in zip(orig_words, corr_words):
        if o.lower() != c.lower():
            result.append(f"{RED}{o}{RESET} → {GREEN}{c}{RESET}")
        else:
            result.append(o)

    return "  |  ".join(result)


test = input("Enter the string: ")

corrected = autocorrect_sentence(test)

print("\nColored Comparison:\n")
print(color_compare(test, corrected))

print("\nFinal Output without GRAMATICAL correction:")
print(corrected)
```
